Выведу уравнение движения:
$$T=\frac{m}{2}\cdot(\dot\alpha \cdot l)^2;П=-m\cdot g \cdot l \cdot cos\alpha $$
Лагранжиан: $$L=T-П=\frac{m}{2}\cdot(\dot\alpha \cdot l)^2+m\cdot g \cdot l \cdot cos\alpha  $$
$$\frac{\partial \frac{\partial L}{\partial \dot\alpha}}{\partial t}-\frac{\partial L}{\partial \alpha}=-k\cdot\dot\alpha $$
$$ m\cdot l^2\cdot\ddot\alpha+m\cdot g\cdot l\cdot sin\alpha=-k\cdot \dot\alpha $$
Добавим управление и поделим на коэффициент при старшей производной:
 $$\ddot\alpha+\kappa\cdot\dot\alpha+\phi\cdot sin\alpha=u$$
 

из условия $$\frac{\partial П}{\partial \alpha}=0 $$
следуют два положения равновесия:
$$\alpha=0;\alpha=\pi$$

## Рассмотрим случай  $$\alpha =0$$


линеаризовав в окрестности нуля:
$$\ddot\alpha+\kappa\cdot\dot\alpha+\phi\cdot \alpha=u$$

Если выход нашей системы такой,что $$y=\alpha$$,
перейдём к вектору x:$$x1=\alpha,x2=\dot\alpha$$
тогда уравнения в терминах представления в пространстве состояний имеют вид:

$$
A=\left(\begin{array}{cc} 
0 & 1\\
-\phi & -\kappa
\end{array}\right)
$$ 

$$
B=\left(\begin{array}{cc} 
0 \\
1
\end{array}\right)
$$ 

$$
C=\left(\begin{array}{cc} 
1\\
0 
\end{array}\right)
$$ 

$$\dot x=A\cdot x+B\cdot u$$

$$y=С^T\cdot x$$

получу передаточную  функцию $$G_{yu}(s)$$

$$x=(sI-A)^{-1}\cdot B\cdot u$$
$$(sI-A)^{-1}=\left(\begin{array}{cc} 
s & -1\\
\phi & s+\kappa
\end{array}\right)^{-1}=\left(\begin{array}{cc} 
s+\kappa & 1\\
-\phi & s
\end{array}\right)\cdot\frac{1}{s^2+\kappa\cdot s+\phi}
$$ 
$$x=\frac{1}{s^2+\kappa\cdot s+\phi}\cdot\left(\begin{array}{cc} 
1\\
s 
\end{array}\right)\cdot u$$

следовательно $$y=\frac{1}{s^2+\kappa\cdot s+\phi}\cdot u$$

$$G_{yu}(s)=\frac{1}{s^2+\kappa\cdot s+\phi}$$

матрица А является устойчивой тк её собственные числа:
    $$\lambda=\frac{-\kappa+_{-}(\kappa^{2}-4\cdot\phi)^{0.5}}{2}$$

действительная часть собственных чисел заведомо меньше 0

следовательно можем использовать, что для устойчивых матриц $$\exp^{At}$$стремиться к 0 при t стремящемся к бесконенчости

и подавая на вход комплексный гармонический сигнал с амплитудой а найдем амплитудно частотную характеристику

$$y(t)=C\cdot \exp^{At}\cdot x(0)+C\cdot \exp^{At}\cdot\int\limits_0^t \exp^{(jwI-A)\tau}\cdot B\cdot ad\tau=\exp^{jwt}\cdot C\cdot (jwI-A)^{-1}\cdot B\cdot a$$

следовательно АЧХ $$W(jw)=G_{yu}(jw)$$

строим годограф:
    $$W(jw)=U(w)+j\cdot V(w)$$

$$W(jw))=\frac{1}{(\phi-w^2)+j\cdot w\cdot k}=\frac{(\phi-w^2)-j\cdot w\cdot k}{(\phi-w^2)^2+ w^2\cdot k^2}$$

$$U(w)=\frac{(\phi-w^2)}{(\phi-w^2)^2+ w^2\cdot k^2}$$

$$V(w)=\frac{- w\cdot k}{(\phi-w^2)^2+ w^2\cdot k^2}$$

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
ax.set_xlabel('U')
ax.set_ylabel('V')
x = np.linspace(0, 100000, 10000000)
f=1
k=1
ax.plot((f-x**2)/((f-x**2)**2+x**2*k**2), -x*k/((f-x**2)**2+x**2*k**2),label = 'k=1');
k=2.5
ax.plot((f-x**2)/((f-x**2)**2+x**2*k**2), -x*k/((f-x**2)**2+x**2*k**2),label = 'k=2.5');
ax.legend()
plt.show()
```

```python
ax = plt.axes()
ax.grid()
ax.set_xlabel('U')
ax.set_ylabel('V')
x = np.linspace(0, 100000, 10000000)
f=1
k=0.1
ax.plot((f-x**2)/((f-x**2)**2+x**2*k**2), -x*k/((f-x**2)**2+x**2*k**2),label = 'k=0.1');
ax.legend()
plt.show()
```

амплитудная и фазовая характеристики

$$A(w)=|W(jw)|=|\frac{(\phi-w^2)-j\cdot w\cdot k}{(\phi-w^2)^2+ w^2\cdot k^2}|=\frac{((\phi-w^2)^2+( w\cdot k)^2)^{0.5}}{(\phi-w^2)^2+ w^2\cdot k^2}=\frac{1}{((\phi-w^2)^2+ w^2\cdot k^2)^{0.5}}$$

$$f(w)=\arctan\frac{-w\cdot k}{\phi-w^2}$$

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
ax.set_xlabel('w')
ax.set_ylabel('A')
plt.suptitle("АЧХ")
x = np.linspace(0, 10, 100)
f=1
k=1
ax.plot(x, 1/((f-x**2)**2+x**2*k**2)**0.5,label = 'k=1');
k=2.5
ax.plot(x, 1/((f-x**2)**2+x**2*k**2)**0.5,label = 'k=2.5');

ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
ax.set_xlabel('w')
ax.set_ylabel('A')
plt.suptitle("АЧХ")
x = np.linspace(0, 10, 100)
f=1
k=0.1
ax.plot(x, 1/((f-x**2)**2+x**2*k**2)**0.5,label = 'k=0.1');

ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
ax.set_xlabel('w')
ax.set_ylabel('f')
plt.suptitle("ФЧХ")
x = np.linspace(0, 8, 100)
f=1
k=1
ax.plot(x, np.arctan(-x*k/(f-x**2)),label = 'k=1');
k=2.5
ax.plot(x, np.arctan(-x*k/(f-x**2)),label = 'k=2.5');

ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
ax.set_xlabel('w')
ax.set_ylabel('f')
plt.suptitle("ФЧХ")
x = np.linspace(0, 8, 100)
f=1

k=0.1
ax.plot(x, np.arctan(-x*k/(f-x**2)),label = 'k=0.1');

ax.legend()
plt.show()
```
